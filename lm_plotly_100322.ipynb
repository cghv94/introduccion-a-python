Activando las librerías

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Importando las bases de datos

```python
btc = pd.read_csv('Bitcoin USD (BTC-USD)2022-03-10.csv')
meta = pd.read_csv('Meta Platforms, Inc.  (FB)2022-03-10.csv')
micro = pd.read_csv('Microsoft Corporation (MSFT)2022-03-10.csv')
```

Creando una tabla con todas las bases de datos

```python
base = pd.concat([meta, btc, micro], axis=0)
base
```

Graficando varias líneas con plotly

```python
tres_lineas = px.line(base, x = 'Fecha', y = 'Cierre', title=' Meta Platforms, Inc & BTC-USD, Microsoft', color = 'Nombre')
tres_lineas.show()
```

Plotly modelo mínimos cuadrados ordinarios

```python
# https://plotly.com/python/ml-regression/
meta_ols = px.scatter(meta, x='Apertura', y='Cierre', trendline="ols", trendline_color_override='red')
meta_ols.show()
```

Regresión lineal con sklearn

```python
x_meta = meta.Apertura.values.reshape(-1, 1)

model_meta = LinearRegression()
model_meta.fit(x_meta, meta.Cierre)

x_meta_range = np.linspace(x_meta.min(), x_meta.max(), 100)
y_meta_range = model_meta.predict(x_meta_range.reshape(-1, 1))

meta_lm = px.scatter(meta, x='Apertura', y='Cierre', opacity=0.65)
meta_lm.add_traces(go.Scatter(x=x_meta_range, y=y_meta_range, name='Regression Fit'))
meta_lm.show()
```

Train-Test modelo lineal generalizado (regresión lineal simple)

```python
X = meta.Apertura[:, None]
X_train, X_test, y_train, y_test = train_test_split(X, meta.Cierre, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

x_range = np.linspace(X.min(), X.max(), 100)
y_range = model.predict(x_range.reshape(-1, 1))

fig = go.Figure([
    go.Scatter(x=X_train.squeeze(), y=y_train, name='train', mode='markers'),
    go.Scatter(x=X_test.squeeze(), y=y_test, name='test', mode='markers'),
    go.Scatter(x=x_range, y=y_range, name='prediction')
])

fig.show()
```
